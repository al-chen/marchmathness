```python
def show_results(submission_csv, slots=[]):
    df = pd.read_csv(submission_csv)
    if len(slots) == 0:
        slots = slots = df['Slot'].unique()

    for slot in slots:
        filtered_df = df[(df['Slot'] == slot)]
        seed_counts = filtered_df['Team'].value_counts()
        print("Most occurrences for slot {}".format(slot))
        for (seed, occurrences) in seed_counts.head(4).items():
            seed_df = seeds_df[(seeds_df['Seed'] == seed)]
            if len(seed_df) > 1:
                print("seed_df contains more than one row")
            else:
                team_id = seed_df.iloc[0]['TeamID']
                team_name = teams_df[teams_df['TeamID'] == team_id].iloc[0]['TeamName']
                print(f"{team_name} wins {occurrences/num_simulations * 100:.2f}% of the time")
        print("")
```

```python
show_results('2025_mov_my_submission.csv')
```
